```python
%pylab inline
import cv2
```

```python
#Generates a black image
width = 256
height = 256
blank_image = np.zeros((height,width), np.uint8)
cv2.imwrite('blackImage.jpg', blank_image) 
plt.imshow(blank_image, cmap="Greys_r")
```

```python
canvas = cv2.imread('blackImage.jpg', -1)
row, col = canvas.shape
mean = 0
gauss = np.random.normal(mean, 100, (row,col))
gauss = gauss.reshape(row, col)
canvas = canvas + gauss
cv2.imwrite('noisy.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread('noisy.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
cv2.circle(canvas, (width/2, height/2), 3, (255, 255, 255), -1) 
cv2.imwrite('sourcePoint.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```

```python
canvas = cv2.imread('sourcePoint.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE) 
canvas = cv2.GaussianBlur(canvas, (3,3), 3, 3)
cv2.imwrite('PSFSourcePoint.jpg', canvas)
plt.imshow(canvas, cmap="Greys_r")
```
